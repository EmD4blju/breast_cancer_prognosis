```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
wpbc = pd.read_csv(
    filepath_or_buffer='../dataset/wpbc_original.csv',
    sep=',',
    header=0
)

wpbc['OUTCOME'].value_counts()
```

```python
wpbc = pd.read_csv(
    filepath_or_buffer='../dataset/wpbc_original_B.csv',
    sep=',',
    header=0
)

(wpbc_train, wpbc_test) = train_test_split(wpbc, test_size=0.2, random_state=42)

train_features = wpbc_train.iloc[:,1:]
train_labels = wpbc_train['OUTCOME']

display(train_features, train_labels)
```

```python
param_grid = {
    'C': np.arange(0,10,1),
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf']
}

search = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)
best_model = search.fit(train_features, train_labels)

print(best_model.best_params_, best_model.best_score_)
```
